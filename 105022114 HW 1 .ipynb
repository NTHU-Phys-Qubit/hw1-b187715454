```python
#  1d  histogram of the end point

import numpy as np 
import matplotlib.pyplot as plt 


Comparators = [0.5, 0.5]   
  
positions = [0] 

dice = np.random.random(300) 

downp = dice < Comparators[0] 
upp = dice > Comparators[1] 

for idownp, iupp in zip(downp, upp): 
    down = idownp and positions[-1] > -1
    up = iupp and positions[-1] < 1
    positions.append(positions[-1] - down + up) 
plt.plot(positions) 
plt.show() 
plt.hist(positions)
```

```python
#  2D  single trajectory
import numpy as np 
import matplotlib.pyplot as plt 

n = 1000

x = np.zeros(n)
y = np.zeros(n) 


for i in range(1,n):
    dice = np.random.random() 
    dice = dice * 2* np.pi
    x[i] = x[i - 1] + np.cos(dice)
    y[i] = y[i - 1] + np.sin(dice)

plt.plot(x, y) 
```

```python
# 2 D end point for an ensemble of trajectories

import numpy as np 
import matplotlib.pyplot as plt 

a = 1000
xf = np.zeros(a)
yf = np.zeros(a)
for i in range(1,a):

    n = 10
    xf[i] = x[n-1]
    yf[i] = y[n-1]
    x = np.zeros(n)
    y = np.zeros(n) 


    for i in range(1,n):
        dice = np.random.random() 
        dice = dice * 2* np.pi
        x[i] = x[i - 1] + np.cos(dice)
        y[i] = y[i - 1] + np.sin(dice)
    
plt.plot(xf,yf,'bo')
```

```python
# 2D histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories

import numpy as np 
import matplotlib.pyplot as plt 

a = 10000
xf = np.zeros(a)
yf = np.zeros(a)
r = np.zeros(a)
for i in range(1,a):

    n = 10
    xf[i] = x[n-1]
    yf[i] = y[n-1]
    r[i] = ( x[n-1] ** 2 + y[n-1] ** 2 ) ** 0.5
    x = np.zeros(n)
    y = np.zeros(n) 


    for i in range(1,n):
        dice = np.random.random() 
        dice = dice * 2* np.pi
        x[i] = x[i - 1] + np.cos(dice)
        y[i] = y[i - 1] + np.sin(dice)
    
plt.hist(r,bins=50)
```
